### Worldwide tweets on ChatGPT

```python
df_country = pd.DataFrame(df[df['country']!='Unknown']['country'].value_counts())
df_country=df_country.reset_index()
df_country = df_country.rename(columns={'index':'Country','country':'Count'})
total = df_country['Count'].sum()
df_country['Percentage']=round(df_country['Count']/total*100,1)
df_country
```

```python
# Load the world map shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))


# Update name for United States in world dataframe
world.loc[world.name == 'United States of America', 'name'] = 'United States'


# Merge the country data onto the GeoDataFrame
world = world.merge(df_country[:10], left_on='name', right_on='Country', how='left')

# Set NaN values to zero
world['Count'] = world['Count'].fillna(0)

# Create the figure and axes objects
fig, ax = plt.subplots(figsize=(20, 8))

# Plot the map
world.plot(column='Count', cmap='OrRd', legend=True, ax=ax, edgecolor='black', linewidth=0.5, vmin=0, vmax=max(world['Count']))

# Add country names and tweet counts as annotations
for index, row in world.iterrows():
    count = int(row['Count'])
    if count > 0:
        plt.annotate(text=f"{row['name']}\n{count}", xy=row['geometry'].centroid.coords[0], ha='center', va='center', size=8)

# Add a legend to the plot
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Set the title and legend
ax.set_title('Number of Tweets by Country', fontsize=16)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show the map
plt.show()
```

#### 2. Tweets in each month

```python
import matplotlib.pyplot as plt

# Group tweet counts by month
tweet_counts_by_month = eda.groupby(pd.Grouper(key='Date', freq='M'))['Tweet'].count()

# Create a figure and set the size
fig = plt.figure(figsize=(10, 6))

# Create bar chart
plt.bar(tweet_counts_by_month.index.strftime('%b %Y'), tweet_counts_by_month.values)

# Set x-axis label rotation
plt.xticks(rotation=45)

# Set chart title and labels
plt.title('Tweets by Month')
plt.xlabel('Month')
plt.ylabel('Number of Tweets')

# Show chart
plt.show()
```

```python
import pandas as pd

eda['Date'] = pd.to_datetime(eda['Date'], format='%d-%m-%Y')
# Set the start date as the first day of the first week
start_date = pd.to_datetime('2022-12-01')

# Create a date range from the first day of the first week to January 31, 2023
dates = pd.date_range(start=start_date, end='2023-02-15')

# create a dictionary to store date-week
date_week = {}
for date in dates:
    week = date.isocalendar()[1] - start_date.isocalendar()[1] + 1
    if week < 0:
        week += 52
    date_week[date] = week
```

```python
eda['Week'] = eda['Date'].map(date_week)
eda['Week'] = eda['Week'].astype('int')
eda['Mon'] = eda['Date'].dt.month_name()
eda['Month'] = eda['Mon'].map({'January': "January,2023", 'December': 'December,2022'})
eda = eda.drop('Mon', axis=1)
```

#### 3. Tweets count in each week

```python
#Count by Week
fig, ax = plt.subplots(1,1, figsize=(8,6))
ax = sns.countplot(x=eda['Week'], palette='PuBuGn')
for p in ax.patches:
    ax.text(p.get_width()/2 + p.get_x(), p.get_height()+1000, p.get_height(), horizontalalignment='center', size =12)
ax.set_title('Count of Tweets by Weeks')
plt.tight_layout()
plt.show()
```

### Active users who discussed about Chat GPT on Twitter

```python
#Tweets frequency
df_user = pd.DataFrame(eda["UserName"].value_counts())
df_user=df_user.reset_index()
df_user = df_user.rename(columns={'index':'User', 'UserName':"Tweets_Frequency"})
bins = [0, 1, 10, 50, 100, df_user["Tweets_Frequency"].max()]
labels=['1', '2-10', '10-50', '50-100', '>100' ]
df_user['Level'] = pd.cut(df_user['Tweets_Frequency'], bins = bins, labels =  labels)
df_user["Level"].value_counts()

fig,ax = plt.subplots(1,2, figsize=(12,5))
plt.subplot(1,2,1)
ax= sns.barplot(x=df_user['Level'].value_counts().index, y = df_user['Level'].value_counts(), palette = 'PuBuGn')
ax.set_ylabel('Counts of Users')
ax.set_xlabel('Number of Tweets')
ax.set_title('User Activity in Discussions about ChatGPT')
for p in ax.patches:
    ax.text(p.get_width()/2+p.get_x(), p.get_height(), int(p.get_height()), ha='center', size=12)
ax.set_yscale('log')

plt.subplot(1,2,2)
ax = sns.barplot(y = df_user['User'][0:10], x=df_user['Tweets_Frequency'][0:10], palette='PuBuGn')
for p in ax.patches:
    ax.text(p.get_width(), p.get_y()+0.5, int(p.get_width()), va='center', size=10)
ax.set_title('Top 10 Most Active ChatGPT Discussing Users')
plt.tight_layout()
plt.show()

total = df_user.shape[0]
once = round(df_user[df_user['Level']=='1'].shape[0]/total*100,2)
print(f'In two and half months (Dec 2022, Jan 2023 and upto 15th Feb 2023),there are a total of {total} \
twitter users tweeted about ChatGPT. \n{once}% users have tweeted only once, while {100-once}% users have tweeted more than once.')
```
