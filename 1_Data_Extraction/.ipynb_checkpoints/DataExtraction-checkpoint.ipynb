## Data extraction

- Data is extracted from twitter using snscrape. 
- All the tweets which has 'chatgpt' in it are extracted from 12/01/2022 - 02/15/2023.

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

query = "chatgpt since:2023-02-10 until:2023-02-12"
tweets = []
limit = 50000

for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    try:
        if len(tweets) == limit:
            break
        elif tweet.lang == 'en':
            tweets.append([tweet.date, tweet.id, tweet.user.username, tweet.user.verified, tweet.content, tweet.user.location, tweet.lang])
    except AttributeError as e:
        # Skip the tweet if there's an attribute error
        print(f"Encountered an error: {e}")
        continue

df1 = pd.DataFrame(tweets, columns=['Date', 'ID', 'User', 'VerifiedStatus', 'Tweet', 'Location', 'Language'])
print(df1)
```
